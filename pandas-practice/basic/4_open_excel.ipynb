```python
import pandas as pd
from pandas import ExcelWriter
from datetime import date, datetime

#create the dataframe
df = pd.DataFrame(
    data=[[date(2025, 6, 3), datetime(2025, 6, 3, 12, 28, 28)]],
    columns=['X', 'Y'],
    index=['Time Record']
)

# writing excel table
with ExcelWriter('/Users/gina/Desktop/output.xlsx') as writer:
    df.to_excel(writer)

```

```python
import zipfile
import os

# Define file paths
excel_file = '/Users/gina/Desktop/output.xlsx'
zip_file = '/Users/gina/Desktop/output.zip'

# Create the ZIP file and add the Excel file
with zipfile.ZipFile(zip_file, 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipf.write(excel_file, arcname=os.path.basename(excel_file))

print("The Excel file has been compressed into output.zip on the Desktop ✅.")
```
