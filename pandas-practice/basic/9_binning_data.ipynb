```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the original CSV file
df = pd.read_csv('/Users/gina/Desktop/SydneyAirQuality.csv')
df.columns =["Date","CO","CO2","NO2","SO2","O3","PM2.5","PM10",	"AQI"]

# Define bin ranges and their corresponding labels
bins = [0, 50, 100, float('inf')]  # three intevals
labels = ["Good", "Moderate Pollution", "Polluted"]

#  Apply binning using pd.cut based on AQI values
df["AQI Category"] = pd.cut(df["AQI"], bins=bins, labels=labels, right=True)

df["AQI Category"].value_counts().plot(kind='bar', title='AQI Category Distribution')
plt.xlabel("AQI Category")
plt.ylabel("Frequency")
plt.show()

# Print the result
print(df)


```
